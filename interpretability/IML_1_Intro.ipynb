# 1 Interpretable Machine Learning

## 1.1 Why ?
Some times it is sufficient to have something that works, without knowing why it works (black box or human brain).

- Just care about performance measures (with test set): MSE, MAE, R-Squared ... 

Other times it is necessary or interesting to know why it works, to a certain degree.

- Desire to increase performance
- Liability (your self driving vehicle killed a person, how did that come, court wants to know)
- Science is often interested in why models work, i.e. how they are representing the world

### Interpretable 

- The dictionary says, capability of being explained or understood, (https://en.wiktionary.org/wiki/interpretable)
- In mathematics interpretable means explainable 
- In logic T is interpretable in S if a translation is possible which conserves all the theorems etc. https://en.wikipedia.org/wiki/Interpretability 
- In ML interpretability can be the degree to which a human can understand the cause of a decision
- In ML interpretability can be the degree to which a human can consistently predict the model's result.

Understand the trained model, i.e. its **structure**, the meaning of the **output**, the meaning and effect of the **fitted weights**, the meaning and impact of the **input** (columns) and their **uncertainties** and impacts.  

### -> Interpretable Machine Learning

## 1.4 Glossary

- Model $$ y = f(\bf{x},\bf{\omega})$$
- y we call **target** (column), (response) function, predictor, output, y, dependent variable, explained variable, regressand, endogonoeus variable ...
- $\bf{x}$ we call **columns, features, inputs**, independent variables, regressors, explanatory variables, observables ...
- $$\hat{y} = f(\hat{\bf{\omega}}|\bf{x})$$ is the estimated/fitted/trained/predicted output. It is based on the fitted parameters. Estimators carry hats.  



## 1.5 Evaluation Metrics

Some much used performance and evaluation metrics in Machine Learning (taken from MV/AM notebook):



1. Regression:
    * Mean Square Error (MSE): $mse=\frac{1}{n}\sum_i(y_i - \hat y(\bar x_i))^2$
    * Mean Absolute Error (MAE): $mae=\frac{1}{n}\sum_i|y_i - \hat y(\bar x_i)|$
    * Median Absolute Deviation (MAD): $mad=median(|y_i - \hat y(\bar x_i)|)$
    * Fraction of the explained variance: $R^2=1-\frac{\sum_i(y_i - \hat y(\bar x_i))^2}{\sum_i(y_i - \bar y_i)^2}$, where $\bar y=\frac{1}{n}\sum_i y_i$

2. Classification:
    * Confusion matrix 

<img src="https://github.com/neworldemancer/DSF5/raw/master/figures/confusion_mtr.png" width="60%"/>

    * Accuracy $=\frac{TP+TN}{TP+FP+FN+TN}$
    * Precision $=\frac{TP}{TP+FP}$
    * Recall $=\frac{TP}{TP+FN}$
    * F1 $=2\frac{Precision \cdot Recall}{Precision+Recall} = \frac{2 TP}{2 TP + FP+FN}$
    * Threat score (TS), or Intersection over Union (IoU): IoU=$\frac{TP}{TP+FN+FP}$
